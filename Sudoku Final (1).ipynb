```python
### Printing like a Sudoku Board Pattern
def print_board(board):
    for row in range(9):
        if row == 3 or row == 6:
            print('- -  - - - - - - - - - -')
        for col in range(9):
            if col == 3 or col == 6:
                print('| ',end=' ')
            print(board[row][col],end = ' ')
        print('\n')
        

### Checking column wise, that is checking whether the value val already occupied in the column.
def col_check(val,col):
    out = 1
    for i in range(9):
        if S[i][col] == val:
            out = 0
    return(out)

### Checking row wise, that is checking whether the value val already occupied in the row.
def row_check(val,row):
    out = 1
    for i in range(9):
        if S[row][i] == val:
            out = 0
    return(out)

### Checking corresponding 3*3 matrix wise, that is checking whether the value val already occupied in the matrix
def matrix_check(val,row,col):
    out = 1
    
    row_ind = int(row/3)
    col_ind = int(col/3)
    for i in range(3):
        for j in range(3):
            r = 3 * row_ind + i
            c = 3 * col_ind + j
            if S[r][c] == val:
                out = 0
    return(out)

### Complete check, that is checking the variable val is a possible solution of the position (row,col) or not
def check(val,row,col):
    out = row_check(val,row)
    out = out * col_check(val,col)
    out = out * matrix_check(val,row,col)
    if out == 1:
        return True
    else:
        return False

### Checking the matrix completely solved or not (or checking if there exist any empty points or not)
def empty(board):
    for i in range(9):
        for j in range(9):
            if board[i][j] == 0:
                return(i,j)
    return False      

### Finding the Indexes of the empty positions
def blank_indx(board):
    out = []
    for i in range(9):
        for j in range(9):
            if board[i][j] == 0:
                out.append()
    return(out)

#### Finding the possible solutions of every empty positions.
def possible_solution(board):
    sol = []
    count = 0
    for i in range(9):
        for j in range(9):
            val = board[i][j]
            if val == 0:
                sol.append([])
                for k in range(1,10):
                    out = check(k,i,j)
                    if out == 1:
                        sol[count].append(k)
                count += 1
    return(sol)

### Normal Mode! Solve upto intermediate level sudoku boards
def Nsolve(board):
    board_copy = board
    while empty_check(board_copy):
        
        sol = possible_solution(board_copy)
        blank = blank_indx(board_copy)
        no_emp = len(sol)
        
        for i in range(no_emp):
            ln = len(sol[i])
            if ln == 1:
                row = blank[i][0]
                col = blank[i][1]
                board_copy[row][col] =sol[i][0]
    
    return(board_copy)

### Active Mode! Solve by recurssive function, solve every sudoku boards
def Asolve():
    global S
    if empty(S) == False:
        return True
    else:
        row,col = empty(S)
    for val in range(1,10):
        if check(val,row,col):
            S[row][col] = val
            if Asolve():
                return True
            else:
                S[row][col] = 0
    return False
        
### Question And Answer
def QandA():
    global S
    print('Question :','\n')
    print_board(S)
    
    Asolve()
    
    print('|||||||||||||||||||||||','\n')
    print('Answer :','\n')
    print_board(S)
```
