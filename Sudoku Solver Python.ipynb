```python
### Printing like a Sudoku Board Pattern
def print_board(board):
    for row in range(9):
        if row == 3 or row == 6:
            print('- -  - - - - - - - - - -')
        for col in range(9):
            if col == 3 or col == 6:
                print('| ',end=' ')
            print(board[row][col],end = ' ')
        print('\n')


### Modified Check function

def check(val,row,col):
    # Checking row wise and Column wise
    for i in range(9):
        if S[row][i] == val:
            return False
        if S[i][col] == val:
            return False
    # Checking square matrix wise
    r = int(row/3)
    c = int(col/3)
    for i in range(3):
        for j in range(3):
            if S[3*r + i][3*c + j] == val:
                return False
    return True

### Checking the matrix completely solved or not (or checking if there exist any empty points or not)
def empty(board):
    for i in range(9):
        for j in range(9):
            if board[i][j] == 0:
                return(i,j)
    return False     

### Solve by recurssive function, solve every sudoku boards
def solve():
    global S
    if empty(S) == False:
        return True
    else:
        row,col = empty(S)
    for val in range(1,10):
        if check(val,row,col):
            S[row][col] = val
            if solve():
                return True
            else:
                S[row][col] = 0
    return False
        
### Question And Answer
def QandA():
    global S
    print('Question :','\n')
    print_board(S)
    
    solve()
    
    print('|||||||||||||||||||||||','\n')
    print('Answer :','\n')
    print_board(S)
```
